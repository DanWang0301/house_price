```
v=pd.read_csv('/content/drive/Shareddrives/machine_learning/house_price/house_price_data/valid_data.csv')
v.columns
```

```
val_condition=v[['sale_yr', 'sale_month', 'bedrooms',
       'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'view',
       'condition', 'grade', 'sqft_above', 'sqft_basement', 'yr_built',
       'yr_renovated', 'zipcode', 'lat', 'long', 'sqft_living15',
       'sqft_lot15']]
val_data=np.array(val_condition).astype(np.float64)

val_price=v[['price']]
val_label=np.array(val_price).astype(np.float64)
```

```
print(val_data.shape,val_label.shape)
```

```
data_train=pd.read_csv('/content/drive/Shareddrives/machine_learning/house_price/house_price_data/train_data.csv')
data_train.columns
```

```
train_text=data_train[['sale_yr', 'sale_month', 'bedrooms',
       'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'view',
       'condition', 'grade', 'sqft_above', 'sqft_basement', 'yr_built',
       'yr_renovated', 'zipcode', 'lat', 'long', 'sqft_living15',
       'sqft_lot15']]
trn_data2=np.array(train_text).astype(np.float64)
print(trn_data2.shape)

train_price=data_train[['price']]
trn_label=np.array(train_price).astype(np.float64)
print(trn_label.shape)
```

```
mean=trn_data2.mean(axis=0)
print(mean)
trn_data3=trn_data2-mean
std=trn_data3.std(axis=0)
print(std)
trn_data3 /= std
print(trn_data3)

val_data2=val_data-mean

val_data2/=std
val_data2.shape
```

```
from keras import models
from keras import layers
from keras.callbacks import EarlyStopping

def build_model():
  model=models.Sequential()
  model.add(layers.Dense(2048, activation='relu',
              input_shape=(trn_data3.shape[1],)))

  model.add(layers.Dense(1024, activation='relu'))
  model.add(layers.Dense(1024, activation='relu'))
  model.add(layers.Dense(768, activation='relu'))
  model.add(layers.Dense(512, activation='relu'))
  
  model.add(layers.Dense(1))
  model.compile(optimizer='adam',loss='mse',metrics=['mae'])
  return model
model=build_model()
```

```
num_epochs=10000
history=model.fit(trn_data3,trn_label,validation_data=(val_data2,val_label),
          epochs=num_epochs,batch_size=128,
          callbacks=[EarlyStopping(monitor='val_mae', patience=25,restore_best_weights=True,verbose=2)])
```

```
print(history.history.keys())
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])



plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['loss', 'val_loss'], loc='upper left') 
plt.show()

plt.plot(history.history['mae'])
plt.plot(history.history['val_mae'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['mae', 'val_mae'], loc='upper left') 
plt.show()
# summarize history for loss plt.plot(history.history['loss']) plt.plot(history.history['val_loss']) plt.title('model loss')
```
